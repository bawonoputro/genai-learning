```python
from transformers import pipeline


#Try different models and prompts
models = ["gpt2", "distilgpt2", "facebook/opt-350m"]
prompts = [
    "Once upon a time",
    "Compared to Asia, Europe",
    "Lately"
]



for model_name in models:
    try:
        print(f"Model: {model_name}")

        generator = pipeline("text-generation", model=model_name)

        for prompt in prompts:
            result = generator(prompt, max_length=60, num_return_sequences=1)
            print(f"\nPrompt: {prompt}")
            print(f"Output: {result[0]['generated_text']}")
    except Exception as e:
        print(f"Failed to load {model_name}: {e}")
        continue


```

```python
#Use previous models but various temperature
#only one prompt

temperature = [0.5, 1.5]
for model_name in models:
    try:
        print(f"Model: {model_name}")

        generator = pipeline("text-generation", model=model_name)

        for temp in temperature:
            result = generator("Tonight, ", max_length=60, num_return_sequences=1,
                               temperature=temp)
            print(f"\nTemperature: {temp}")
            print(f"Output: {result[0]['generated_text']}")
    except Exception as e:
        print(f"Failed to load {model_name}: {e}")
        continue
```

```python
#Use previous models but various top_k
#only one prompt

top_k = [1, 5]
for model_name in models:
    try:
        print(f"Model: {model_name}")

        generator = pipeline("text-generation", model=model_name)

        for k in top_k:
            result = generator("After the sunset, the kid", max_length=60, num_return_sequences=1,
                               top_k=k)
            print(f"\nTop_K: {k}")
            print(f"Output: {result[0]['generated_text']}")
    except Exception as e:
        print(f"Failed to load {model_name}: {e}")
        continue
```
